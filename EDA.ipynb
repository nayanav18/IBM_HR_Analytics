# Conclusions

### 1. BusinessTravel
- **Observation:** Employees who travel "Frequently" show higher attrition than "Rarely" or "Non-Travel".
- **Conclusion:** Heavy travel may increase burnout; employees with frequent travel are at higher risk of leaving.

### 2. Department
- **Observation:** Sales and Research & Development tend to have slightly higher attrition than HR.
- **Conclusion:** Some departments face higher turnover â€” retention programs may be needed in Sales and R&D.

### 3. EducationField
- **Observation:** Fields like "Life Sciences" and "Technical Degree" have slightly higher attrition; others like "Marketing" may show lower.
- **Conclusion:** Certain education backgrounds may feel mismatched with roles, contributing to resignations.

### 4. Gender
- **Observation:** Slightly more males leave than females, but distribution is close.
- **Conclusion:** Gender is not a strong predictor in this dataset; other factors matter more.

### 5. JobRole
- **Observation:** Roles like Sales Executive and Research Scientist have higher attrition; Managers have very low attrition.
- **Conclusion:** Entry-level or non-managerial roles are more prone to leaving; career growth programs could help.

### 6. MaritalStatus
- **Observation:** Single employees leave more frequently than married or divorced.
- **Conclusion:** Singles may be more mobile and likely to switch jobs; retention strategies may differ by marital status.

### 7. OverTime
- **Observation:** Employees working overtime ("Yes") have much higher attrition than those who donâ€™t.
- **Conclusion:** Overtime is a strong driver of attrition â€” work-life balance improvements are essential.
